```python
#As '0' here indicates that previous creditline was faulty in payments or loan was not ever taken,
#Its better to impute with '0' as our model should reject that risky loan before hand ,otherwise underwriter would have take timejust to rejcet application

#Replacing the NaN in Credit_Historywith 0
loan_data.loc[loan_data['Credit_History'].isnull()==True,'Credit_History']=0

#Cheking for changes now
loan_data['Credit_History'].isnull().sum()
```

<p style="width: 700px;padding: 20px;background: papayawhip;border-radius:10px;font-size:17px;border-radius:20px">Now, handling the Categorical dtype
    
   </p>

```python
#Using the label Encoder to convert Categorical data into numerical/binary(in some cases), for the model building

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder() #Creating object

#Now using le.fit_transform

loan_data['Gender'] =le.fit_transform(loan_data['Gender'])
loan_data['Married']=le.fit_transform(loan_data['Married'])
loan_data['Education']=le.fit_transform(loan_data['Education'])
loan_data['Property_Area']=le.fit_transform(loan_data['Property_Area'])
loan_data['Loan_Status (Approved)']=le.fit_transform(loan_data['Loan_Status (Approved)']) # Our target feature
loan_data['Dependents']=le.fit_transform(loan_data['Dependents'])
loan_data['Self_Employed']=le.fit_transform(loan_data['Self_Employed'])
```

```python
#Finanlly checking loan_data
loan_data.info(), 
loan_data.head()
```

<p style="width: 700px;padding: 20px;background: papayawhip;border-radius:10px;font-size:17px;border-radius:20px">Now, Scaling the data
    
   </p>

```python
#As scaling need to be done only for continous data
from sklearn.preprocessing import MinMaxScaler #or use here X_Scale=(X-Xmin)/(Xmax-Xmin) 
scale=MinMaxScaler() #creating the object, toconvert into Standard Normal Distribution

loan_data[['ApplicantIncome','CoapplicantIncome','LoanAmount']] = scale.fit_transform(loan_data[['ApplicantIncome','CoapplicantIncome','LoanAmount']])

#Cheking for transformed data
loan_data.head()
```

<p style="width: 700px;padding: 20px;background: papayawhip;border-radius:10px;font-size:17px;border-radius:20px">Now, Feature Selection
    
   </p>

```python
#Removing the redundant column
loan_data.drop('Loan_ID', axis=1,inplace=True)
#loan_data.head()
```

```python
#Checking for highly correlated features
corr_data= loan_data[['ApplicantIncome','CoapplicantIncome','LoanAmount']]

#Plotting heatmap  to see correlation
sns.heatmap(corr_data.corr(), annot=True)

#There are no string relationship among the sacled numerical data
corr_data.describe()
```

<p style="width: 700px;padding: 20px;background: papayawhip;border-radius:10px;font-size:21px;border-radius:20px">
    <b>Model Creation</b>
    
   </p>

```python
#Defining the Dependent and Target variables
X = loan_data.iloc[:,:-1]
y=loan_data['Loan_Status (Approved)']
```

```python
#Creating Training and Testing variables
from sklearn.model_selection import train_test_split

X_train,X_test,y_train, y_test =train_test_split(X, y, random_state=3)
```

```python
#Balancing the training data to reduce biasness of output towards majority classs
loan_data['Loan_Status (Approved)'].value_counts()
```

```python
#instaling the imblearn package
!pip install imblearn
from imblearn.over_sampling import SMOTE

#Creating object
smote=SMOTE()
```

```python
X_smote, y_smote = smote.fit_resample(X_train, y_train)

#Checking to balance
from collections import Counter
print('Actual Classes', Counter(y_train))
print('SMOTE Classes', Counter(y_smote))
```

```python
#SVM Classifier
from sklearn.svm import SVC #SVC is Support Vector Classifier
svc_base = SVC() #base model with default parameters, creating object

#Fitting the training data
svc_base.fit(X_smote, y_smote)
```

```python
#Predicting the output for X_test
y_hat=svc_base.predict(X_test)
y_hat
```

<p style="width: 700px;padding: 20px;background: papayawhip;border-radius:10px;font-size:17px;border-radius:20px">Evaluation of Base model
    
   </p>

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, classification_report, f1_score
acc= accuracy_score(y_test, y_hat)
print(acc)
print(classification_report(y_test, y_hat))

#recall of 0 in not okay
```

```python
#Confusion matrix
cm=pd.crosstab(y_test, y_hat)
cm
```

<p style="width: 700px;padding: 20px;background: papayawhip;border-radius:10px;font-size:17px;border-radius:20px">Checking for Cross Validation Score
    
   </p>

```python
from sklearn.model_selection import cross_val_score

cval=cross_val_score(svc_base, X, y, cv=5, scoring='f1')
print(cval)
print("Cross Validation Score", cval.mean())
print("std", cval.std())

#std of <0.05 is good
```

<p style="width: 700px;padding: 20px;background: papayawhip;border-radius:10px;font-size:17px;border-radius:20px">
Model Hyperparameter tunning
   </p>

```python
#GrideSearch
from sklearn.model_selection import GridSearchCV

#Defining the parameters range
param_grid ={'C':[0.1,1,10,100,1000],
            'gamma':[1,0.1, 0.01,0.001,0.0001],
             'random_state':(list(range(1,20)))}

model=SVC()
grid_search = GridSearchCV(model,param_grid, refit=True,verbose=2, scoring='f1', cv=10 )

#Fitting the model for GridSearch
grid_search.fit(X,y) #Use 
```

```python
#Printing best params after tunning
print(grid_search.best_params_)
```

```python
#Priniting how our model looks after HyperParameter tunning
print(grid_search.best_estimator_)
```

```python
#clf=SVC(C=100, gamma=0.001, random_state=42) ##0.1
svc_clf= SVC(C=10, gamma=0.01, random_state=1) ##0.1
```

```python
#Fitting svc_clf
svc_clf.fit(X_smote,y_smote)

#Predicitng svc_clf
y_clf = svc_clf.predict(X_test)
```

```python
#Prinitng classification report on svc_clf
print(classification_report(y_test, y_clf))
```

```python
#Confusion matrix
cm=pd.crosstab(y_test,y_clf)
cm
```

```python
print(f1_score(y_test, y_clf))
```

```python
#Scores after GridSearch
scores_clf = cross_val_score(svc_clf, X,y,cv=10, scoring='f1')
print(scores_clf)
print("Cross Validation Score:", scores_clf.mean())
print("Std:", scores_clf.std()) #std of <0.05 is good
```
