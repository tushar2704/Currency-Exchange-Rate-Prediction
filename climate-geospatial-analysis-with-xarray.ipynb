<p style="width: 700px;padding: 20px;background: papayawhip;border-radius:10px;font-size:17px;border-radius:20px">Applying grouped operations over multidimensional data</p>


<p style="width: 700px;padding: 20px;background: papayawhip;border-radius:10px;font-size:17px;border-radius:20px">Groupby</p>

```python
(data.groupby(data.time.dt.year).mean(dim=['time','latitude', 'longitude']).skt.plot())
```

```python
#by months
(data.groupby(data.time.dt.month).mean(dim=['time','latitude', 'longitude']).skt.plot())
```

```python
# for vegetation
(data.groupby(data.time.dt.month).mean(dim=['time','latitude', 'longitude']).lai_hv.plot())
```

<p style="width: 700px;padding: 20px;background: papayawhip;border-radius:10px;font-size:17px;border-radius:20px">Resample</p>

```python
(data.resample({'time':"10y"}).mean(dim=['time','latitude','longitude']).skt.plot())
```

<p style="width: 700px;padding: 20px;background: papayawhip;border-radius:10px;font-size:17px;border-radius:20px">Merging and concatenating datasets</p>

```python
#years array only
year= data.time.dt.year
year
```

<p style="width: 700px;padding: 20px;background: papayawhip;border-radius:10px;font-size:17px;border-radius:20px">Merging </p>

```python
merge_1 = data.merge(year)
merge_1 #now yearvariable is added with only time dimension
```

```python
#mergewith squared dist with longitude and latitude coordinates

da= data.longitude**2+data.latitude**2
da.name="squared_distance"
merge_2 = data.merge(da)
merge_2
```

```python
#Everything together
merge_3 = merge_1.merge(merge_2)
merge_3
```

<p style="width: 700px;padding: 20px;background: papayawhip;border-radius:10px;font-size:17px;border-radius:20px">Concatenating datasets</p>

```python
#Concat with new dimension dim_power

df_sq_root = merge_3**(1/2)
df_cu_root = merge_3**(1/3)

concat_df = xr.concat([merge_3,df_sq_root, df_cu_root],"dim_power")
concat_df
```

```python
#Plotting tp(total precipitation) onoriginal data
data.sel(time='2019-01-01').tp.plot()
```

```python
concat_df.sel(dim_power=0, time='2019-01-01').tp.plot()
```

```python
#Square root of precipitation
concat_df.sel(dim_power=1, time='2019-01-01').tp.plot()
```

```python
#Cube root of total precipation
concat_df.sel(dim_power=2, time='2019-01-01').tp.plot()
```
