```python
#Data manipulation imports
import numpy as np
import pandas as pd

#Data Visualization imports
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
pio.templates.default = 'plotly_white'

#Additional imports
from math import sqrt
import warnings
warnings.simplefilter(action="ignore", category=FutureWarning)
```

<p style="width: 1000px;padding: 20px;background: papayawhip;border-radius:10px;font-size:25px;border-radius:20px">
    Loading and checking data
</p>

```python
data =pd.read_csv("/kaggle/input/uk-monthly-gdp-2022/UK_monthly_gdp.csv", index_col=None)
```

<p style="width: 1000px;padding: 20px;background: papayawhip;border-radius:10px;font-size:25px;border-radius:20px">
    GDP growth over time
</p>

```python
fig =go.Figure(data=go.Heatmap(z=[data['GDP Growth']], x=data.index, y=['GDP Growth'], colorscale='viridis'))

fig.update_layout(title='GDP Growth', xaxis_title='Time Period', yaxis_title='')

fig.show()
```

```python
#Converting monthly data into quaterly data by resampling
data['Time Period']=pd.to_datetime(data['Time Period'], format='/%m/%Y')

data.set_index("Time Period", inplace=True)
qtr_data = data.resample('Q').mean()
print(qtr_data.head())
```

```python
# Calculating recession based on quarterly GDP growth
qtr_data['Recession'] = ((qtr_data['GDP Growth']<0) & (qtr_data['GDP Growth'].shift(1)<0))

# Filling missing values with False (since the first quarter cannot be in a recession)
qtr_data['Recession'].fillna(False, inplace=True)

# Plotting the GDP growth and recession data
fig = go.Figure()

fig.add_trace(go.Scatter(x=qtr_data.index, y=qtr_data['GDP Growth'], name='GDP Growth',line=dict(color='green', width=5)))

fig.add_trace(go.Scatter(x=qtr_data[qtr_data['Recession']].index, y=qtr_data[qtr_data['Recession']]['GDP Growth'],
                         name='Recession',line=dict(color='red', width=5)))

fig.update_layout(title='GDP Growth and Recession over Time (Quarterly Data)',
                  xaxis_title='Time Period',
                  yaxis_title='GDP Growth')
fig.show()
```

<p style="width: 1000px;padding: 20px;background: papayawhip;border-radius:10px;font-size:25px;border-radius:20px">
    Severity of the Recession
</p>


```python
qtr_data['Recession start'] =qtr_data['Recession'].ne(qtr_data['Recession'].shift()).cumsum()

recession_periods = qtr_data.groupby('Recession start')
recession_duration = recession_periods.size()
recession_severity = recession_periods['GDP Growth'].sum()

fig = go.Figure()
fig.add_trace(go.Bar(x=recession_duration.index, y=recession_duration,
                     name='Recession Duration'))
fig.add_trace(go.Bar(x=recession_severity.index, y=recession_severity,
                     name='Recession Severity'))

fig.update_layout(title='Duration and Severity of Recession',
                  xaxis_title='Recession Periods',
                  yaxis_title='Duration/Severity')

fig.show()
```
